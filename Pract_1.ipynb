# Activation Functions

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 400)
def linear_activation(x):
    return x
x_values = range(-5, 5)
y_values = [linear_activation(x) for x in x_values]
plt.figure(figsize=(6, 6))
plt.subplot(2, 2, 1)
plt.plot(x_values, y_values)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Linear Activation Function')
plt.grid(True)



def sigmoid(x):
    return 1 / (1 + np.exp(-x))
plt.subplot(2, 2, 2)
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)

def relu(x):
    return np.maximum(0, x)
plt.subplot(2, 2, 3)
plt.plot(x, relu(x), label='ReLU', color='orange')
plt.title('ReLU Activation Function')
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)


# Tanh Function
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.subplot(2, 2, 4)
plt.plot(x, tanh(x), label='Tanh', color='green')
plt.title('Tanh Activation Function')
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)
plt.show()

# Leaky ReLU Function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
plt.figure(figsize = (3,3))
plt.plot(x, leaky_relu(x), label='Leaky ReLU', color='red')
plt.title('Leaky ReLU Activation Function')
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.grid(True)
plt.show()


#Softmax Function
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)
plt.figure(figsize =(3,3))
plt.plot(x, softmax(x), label='Softmax', color='blue')
plt.title('Softmax Activation Function')
plt.xlabel("x")
plt.ylabel("softmax(x)")
plt.grid(True)
plt.show()


# ELU Function
def elu(x, alpha=0.5):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
plt.figure(figsize=(3,3))
plt.plot(x, elu(x), label='ELU', color='yellow')
plt.title('ELU Activation Function')
plt.xlabel("x")
plt.ylabel("elu(x)")
plt.grid(True)
plt.show()


# Softplus Function
def softplus(x):
    return np.log(1 + np.exp(x))
plt.figure(figsize= (3,3))
plt.plot(x, softplus(x), label='Softplus', color='purple')
plt.title('Softplus Activation Function')
plt.xlabel("x")
plt.ylabel("softplus(x)")
plt.grid(True)
plt.show()




# SiLU (Sigmoid Linear Unit) Function
def silu(x):
    return x / (1 + np.exp(-x))
plt.figure(figsize=(3,3))
plt.plot(x, silu(x), label='SiLU', color='brown')
plt.title('SiLU Activation Function')
plt.xlabel("x")
plt.ylabel("silu(x)")
plt.grid(True)
plt.show()
